```python
def concept_to_proposed(concept_tsv_path):
    row_number = 1
    concepts = []
    
    def is_id(s):
        try:
            int(s)
            return True
        except ValueError:
            return False
            
    for row in csv2list(concept_tsv_path, comment='!!!')[1:]:
        row_number += 1
        
        if is_id(row[0]):
            concepts.append(
                ProposedConcept(row[0], row_number)
            )
        
    return concepts
```
